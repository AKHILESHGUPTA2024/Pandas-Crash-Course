### Pandas Quick Tutorial Overview

###### Pandas is an open source library used in data extraction, data analytics, data manipulation, data cleaning, data transformations, it contains Series and DataFrame

##### Series : like an individual column of a table

```python
import pandas as pd #pd is the short name of pandas as widely used with this name, hence follow the same naming convention
data = ['apple','banana','pineapple','watermelon','grapes']
Ser1 = pd.Series(data,name='fruits')
Ser2 = pd.Series([1,2,3,4,5],index=['a','b','c','d','e'])
Ser2
```

```python
# you can also describe some categorical columns as
df[['Gender','Customer ID']].describe()
```

```python
#sub setting columns
df[['Gender','Married']] # taken 2 columns
df[['Customer ID', 'Gender', 'Age', 'Married', 'Number of Dependents',
       'City', 'Zip Code', 'Latitude', 'Longitude', 'Number of Referrals']] # taken 10 columns
df[df.columns[:5]] # this will show 5 columns start with 0 till 5 rows
```

```python
# taking samples
df.sample(5) # taken 5 rows based on overall dataframe
df.sample(5,random_state=500) # taken 5 rows based on 500 rows of the dataframe
df.sample(frac=0.1) # taken 10% sample based on overall dataframe
df.sample(frac=0.001, random_state=242) # taken a fraction sample based on random 242 rows
```

```python
# filtering using loc | iloc (loc = for columns & iloc = for rows)
df.loc[df['Gender']=='Male'] # filter only Males
df.loc[(df['Gender']=='Male') & (df['Age']>50)] # filter only Males whose Ages are >50
df.loc[~(df['Gender']=='Male') & (df['Age']>50)] # reverse of the above using ~ sign
df.loc[:,['Customer ID', 'Gender','Married']] # this will show all rows for the specified column names (in this case - 3 columns)

df.iloc[:5]#this will show top 5 rows
df.iloc[100:200] #this will show rows started from 100 till 200
df.iloc[0,5] #this will show 1st row and 5th col value (single value)
df.iloc[:5,:5] # this will show 5 rows start from 0 till 5th and columns start from 0 till 5th
df.iloc[[5]] # this will show only 5th row
df.iloc[:,5] # this will show rows start from 0 till end with only 5th column
```

```python
#select datatypes
df.select_dtypes('int')#only numerical columns will be displayed, you can change datatypes if you wish to
```

```python
#Query
df.query('(Gender=="Male") and Age>50')
```

```python
# summary statistics
print(df['Age'].min())
print(df['Age'].max())
print(df['Age'].mean())
print(df['Age'].std())
print(df['Age'].sum())
print(df['Age'].quantile(0.5))
print(df['Age'].quantile([0.25,0.75]))
```

```python
#aggregations
df[['Age','Tenure in Months','Latitude','Avg Monthly Long Distance Charges','Longitude']].agg(['mean','sum','max','std'])# 1st way
df[['Age','Tenure in Months','Latitude','Avg Monthly Long Distance Charges','Longitude']].agg(                           #2nd way
    {'Age':['mean','sum'],
     'Tenure in Months':['max','min'],
     'Latitude':['max','min'],
     'Avg Monthly Long Distance Charges':['mean','max'],
    
     'Longitude':['mean','sum','max','std']})
```

```python
#summarize categorically
print(df['Gender'].unique()) # display only unique values or categories
print(df['Gender'].nunique()) # display only unique numbers 
df['Gender'].value_counts() # display numbers categorically
df[['Gender','Married']].value_counts() # display numbers categorically (with multiple columns)
df[['City','Internet Type']].value_counts().reset_index() # this way index number got reset
```
